## Q17. List the most frequent routes(same START & END points) taken by Uber drivers. 
## (3 points)

## Q22. What is proportion of trips that is Business and what is the proportion of trips that is Personal? (3 points)

```python
biz_count=df[df['CATEGORY*']=="Business"].count()
per_count=df[df['CATEGORY*']=="Personal"].count()
total=biz_count["CATEGORY*"]+per_count["CATEGORY*"]

Prop_biz_trips=biz_count['CATEGORY*']/total
print("Propotion of Business Trips is :",Prop_biz_trips*100,"%")
print("Propotion of Personal Trips is :",(1-Prop_biz_trips)*100,"%")
```
