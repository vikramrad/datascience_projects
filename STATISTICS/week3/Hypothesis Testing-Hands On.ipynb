```python
import numpy as np
import pandas as pd
from scipy.stats import ttest_1samp, ttest_ind
```

```python
mydata = pd.read_csv('CardioGoodFitness.csv')
```

```python
mydata.head()
```

```python
mydata.info()
```

```python
mydata.describe(include="all")
```

```python
mydata.mean()
```

```python
t_statistic, p_value = ttest_1samp(mydata['Fitness'], 3)
print(t_statistic,p_value)
```

```python
(np.mean(mydata['Fitness'] - 3)) / (np.std(mydata['Fitness'],ddof=1)/np.sqrt(180))
```

```python
print ("one-sample t-test p-value=", p_value)
```

```python
#Two sample testing

mydata.groupby(['Fitness']).groups.keys()
```

```python
mydata.groupby('Fitness').first()
```

```python
group1=mydata[mydata.Gender=="Male"].Fitness
```

```python
group2=mydata[mydata.Gender=="Female"].Fitness
```

```python
t_statistic, p_value = ttest_ind(group1, group2)
print(t_statistic,p_value)
```
